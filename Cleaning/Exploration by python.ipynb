```python
import pandas as pd

file_path = 'Data_Modified.csv'
data = pd.read_csv(file_path)

state_category_sales = data.groupby(['State', 'Category']).size().reset_index(name='Count')

most_sold_category_by_state = state_category_sales.loc[state_category_sales.groupby('State')['Count'].idxmax()]

most_sold_category_by_state = most_sold_category_by_state[['State', 'Category', 'Count']]

most_sold_category_by_state
```

```python
# Group by Category and sum sales
category_sales = data.groupby('Category')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
category_sales
```

```python
# Convert Order Date to datetime and extract year
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Year'] = data['Order Date'].dt.year

# Group by Year and sum sales
sales_by_year = data.groupby('Year')['Sales'].sum().reset_index()
sales_by_year
```

```python
# Group by Product Name and sum sales, then sort in descending order
most_sold_products = data.groupby('Product Name')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
most_sold_products.head(10)  # Show top 10
```

```python
# Same as above but sorted in ascending order
least_sold_products = data.groupby('Product Name')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=True)
least_sold_products.head(10)  # Show bottom 10
```

```python
# Group by Country and Region, sum sales
country_region_sales = data.groupby(['Country', 'Region'])['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
country_region_sales
```

```python
# Group by City and sum sales
city_sales = data.groupby('City')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
city_sales.head(10)  # Show top 10
```

```python
##2. Performance Questions by Region and Category
```

```python
# Group by Region and Category, sum sales
region_category_sales = data.groupby(['Region', 'Category'])['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
region_category_sales
```

```python
# Group by City and State, sum sales
city_state_sales = data.groupby(['State', 'City'])['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
city_state_sales
```

```python
## 3. Questions about Customers
```

```python
# Group by Customer Segment and sum sales
segment_sales = data.groupby('Segment')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
segment_sales
```

```python
# Group by Region and Customer Segment, sum sales
region_segment_sales = data.groupby(['Region', 'Segment'])['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
region_segment_sales
```

```python
# Group by Customer ID and count the number of orders
most_valuable_customers = data.groupby('Customer ID')['Order ID'].nunique().reset_index(name='Order Count').sort_values(by='Order Count', ascending=False)
most_valuable_customers.head(10)  # Show top 10
```

```python
## 4. Shipping Questions
```

```python
# Calculate shipping time and get the average
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data['Shipping Time'] = (data['Ship Date'] - data['Order Date']).dt.days

average_shipping_time = data['Shipping Time'].mean()
average_shipping_time
```

```python
shipping_mode_sales = data.groupby('Ship Mode')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
shipping_mode_sales
```

```python
# Group by Ship Mode and Category, sum sales
shipping_category_sales = data.groupby(['Ship Mode', 'Category'])['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
shipping_category_sales
```
